```python
from gensim.models.phrases import Phraser
from gensim.models import Phrases
from nltk.corpus import stopwords
from google.colab import drive
import pandas as pd
import numpy as np
import spacy
import json
import re
import os

# mount Google Drive to access the uploaded data
drive.mount('/content/drive')
dir = '/content/drive/MyDrive/Panasonic/'
os.chdir(dir)
```

```python
# load Spacy's large NLP model
nlp = spacy.load('en_core_web_lg')
```

```python
def clean(s):
    # lowercase
    s = s.lower()
    
    # remove chars that are:
    #   ^\w: NOT word char (not alphanumeric)
    #   ^\s: NOT space
    #   ^. : NOT period
    #   |_ : OR underscore
    s = re.sub(r'[^.\w\s]|_', ' ', s)
    
    # remove excess spaces
    s = re.sub(' +', ' ', s)
    s = '\n'.join(line.strip().rstrip('.') for line in s.split('\n'))

    return s
```

```python
def print_header(header):
    print('+-' + len(header)*'-' + '-+')
    print(f'| {header} |')
    print('+-' + len(header)*'-' + '-+')
```

```python
def get_orgs(file):
    with open(file, 'r') as f:
        url = f.readline().strip()
        text = clean(f.read().strip())
        doc = nlp(text)
        orgs = []
        for ent in doc.ents:
            if ent.label_ == 'ORG':
                org = ent.text.strip()
                if org not in orgs:
                    orgs.append(org)
        return url, orgs
```

```python
folder = 'saved_data'
companies = ['Aisle411']
#companies = os.listdir(folder)
result = {}
for company in companies:
    print(company)
    file_number = 0
    while 1:
        path = os.path.join(folder, company, str(file_number) + '.txt')
        if os.path.exists(path):    
            print(path)
            url, orgs = get_orgs(path)
            result[url] = orgs
            file_number += 1
        else:
            break
```

```python
for url in result:
    print_header(url)
    for org in result[url]:
        print(org)
```
